```python
import pandas as pd
# чтоб картинки рисовались в тетрадке
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10, 8)
```

```python
data = pd.read_csv('../../data/adult.data.csv')
data.head()
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
import numpy as np
data.groupby('sex')[['sex']].agg('count').head()
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
data[data['sex'] == 'Female']['age'].mean()
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
data['native-country'].value_counts(normalize=True)['Germany']
```

**4. Постройте гистограмму распределения (*bar plot*) образования людей (признак *education*).**

```python
edu_frame = data['education'].value_counts()
edu_data = edu_frame.values
n_range = range(len(edu_data))
plt.bar(n_range, edu_data)
plt.xticks(n_range, edu_frame.index)
```

**5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
import numpy as np
data.groupby(['salary'])[['age']].agg([np.mean, np.std])
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education - Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
high_education_levels = ('Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate')
not data[(data.salary == '>50K') & ~data.education.isin(high_education_levels)].age.count()
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
data.groupby(['race', 'sex'])[['age']].describe()
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
married_criteria = ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']
men_data = data[data['sex'] == 'Male']
married = men_data['marital-status'].isin(married_criteria)
pd.crosstab(married, men_data['salary'], normalize=True)
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
data[data['hours-per-week'] == data['hours-per-week'].max()].salary.value_counts(normalize=True)
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
data.pivot_table(['hours-per-week'], ['native-country', 'salary'], aggfunc='mean')
```
